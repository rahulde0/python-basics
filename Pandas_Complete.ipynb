## Group By

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
# We want to find average spending group by size of the group.
tips.groupby('size')[['total_bill','tip']].mean()
```

```python
tips.groupby(['size'])[['total_bill','tip']].agg(['mean', 'count'])
```

```python
df_1 = tips.groupby(['sex', 'time'])[['total_bill','tip']].agg(['mean', 'count'])
```

```python
df_1
```

```python
# Return the value for mean tip where Female Paid for Dinner
#Return the #no of times where Female Paid for Dinner
# df_1.loc[('Female', 'Dinner'), [('total_bill','mean'), ('tip':'mean')]]

# Corrected code to select specific columns from a MultiIndex DataFrame
df_1.loc[('Female', 'Dinner'), [('total_bill', 'mean'), ('total_bill', 'count')]]
```

```python
# Sort the Group by DataFrame
tips.groupby(['size'])[['total_bill','tip']].agg(['mean', 'count'])\
.sort_values(by=('total_bill', 'count'), ascending=False) # Single Sorting sequence.
```

```python
# We use multiple sorting sequences to handle clash
tips.groupby(['size'])[['total_bill','tip']].agg(['mean', 'count'])\
.sort_values(by=[('total_bill', 'count'), ('tip','mean')], ascending=[False, False] )
```

```python
import pandas as pd
air_df = pd.read_csv('https://raw.githubusercontent.com/amity1415/eKeeda/refs/heads/main/Rawdata/AQI.csv')
air_df.head()
```

```python
air_df.describe() # Returns the Statistical Summmary of Numerical columns in the data set.
```

```python
air_df.info() # Currently Date and Time are being considered as string objects.
#In order to tell python, hey!! they are Date and Time, and we need to perform Date/time operations on them, we will need to convert
# them to DateTime in Python
```

```python
air_df.head(1)
```

```python
# Instructing Python to consider the Date column as a Python DateTime object.
air_df['Date'] = pd.to_datetime(air_df['Date'], format='%m/%d/%Y') # generic time stamp format = '%m/%d/%Y %H:%M:%S'
air_df['Time'] = pd.to_datetime(air_df['Time'], format= '%H:%M:%S')
```

```python
air_df.head(1)
```

```python
air_df['Date'].astype(str) + ' ' + pd.to_datetime(air_df['Time']).dt.time.astype(str)
```

```python
# Combine the Date and the Time column to give a unified Date-Time Stamp
date_time = pd.to_datetime(air_df['Date'].astype(str) + ' ' + pd.to_datetime(air_df['Time']).dt.time.astype(str))
```

```python
air_df.insert(0, 'Date_Time', date_time)
air_df.head(1)
```

```python
air_df['Date_Time'][:5]
```
