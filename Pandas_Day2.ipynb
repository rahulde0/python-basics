## Series
These are 1d labelled ndarray. capable of storing any data type(int, float, string)
* The axis labels are called index. --> They need not be unique but must be hashable type. (immutable).
* An object is hashable if it has a hash valye which never changes during its lifetime.
* All of Python's immutable built in objects are hashable, while mutable containers(lists, dictionary,e.t.c shall not be used as index)

## Group By

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
# We want to find average spending group by size of the group.
tips.groupby('size')[['total_bill','tip']].mean()
```

```python
tips.groupby(['size'])[['total_bill','tip']].agg(['mean', 'count'])
```

```python
df_1 = tips.groupby(['sex', 'time'])[['total_bill','tip']].agg(['mean', 'count'])
```

```python
df_1
```

```python
# Return the value for mean tip where Female Paid for Dinner
#Return the #no of times where Female Paid for Dinner
# df_1.loc[('Female', 'Dinner'), [('total_bill','mean'), ('tip':'mean')]]

# Corrected code to select specific columns from a MultiIndex DataFrame
df_1.loc[('Female', 'Dinner'), [('total_bill', 'mean'), ('total_bill', 'count')]]
```

```python
# Sort the Group by DataFrame
tips.groupby(['size'])[['total_bill','tip']].agg(['mean', 'count'])\
.sort_values(by=('total_bill', 'count'), ascending=False) # Single Sorting sequence.
```

```python
# We use multiple sorting sequences to handle clash
tips.groupby(['size'])[['total_bill','tip']].agg(['mean', 'count'])\
.sort_values(by=[('total_bill', 'count'), ('tip','mean')], ascending=[False, False] )
```

# #Merging, Concatening

```python
#Concat
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']},
                 index=[0,1,2,3])
df2=pd.DataFrame({'E':['E2','E3','E4','E5'],
                  'F':['F2','F3','F4','F5'],
                  'C':['C2','C3','C4','C5'],
                  'D':['D2','D3','D4','D5']},
                 index=[2,3,4,5])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], axis=0)
```

```python
print(df1)
print('----------')
print(df2)
```

```python
#Merge in DataFrame
pd.merge(left = df1, right=df2, how='inner', on='C')
```

```python
pd.merge(left = df1, right=df2, how='left', on='C')
```

```python
pd.merge(left = df1, right=df2, how='right', on='C')
```

```python
pd.merge(left = df1, right=df2, how='outer', on='C')
```

```python
pd.merge(left = df1, right=df2, how='outer', on=['C','D'])
```

```python
pd.merge(left = df1, right=df2, how='inner', on=['C','D'])
```
