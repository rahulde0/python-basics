## Series
These are 1d labelled ndarray. capable of storing any data type(int, float, string)
* The axis labels are called index. --> They need not be unique but must be hashable type. (immutable).
* An object is hashable if it has a hash valye which never changes during its lifetime.
* All of Python's immutable built in objects are hashable, while mutable containers(lists, dictionary,e.t.c shall not be used as index)

```python
arr =np.array([1,2,3,4,5])
pd_s = pd.Series(arr)
pd_s
```

## DataFrames
DataFrames are 2D data structures having data aligned in tabular forma.

* Data is aligned in rows( index), and columns and can store any data types like int, string , float, boolean.


```python
#Create an empty Data Frame
df = pd.DataFrame()
print(df)
```

```python
fruits = ['Apple','Banana','Mango','Grapes']
pd.DataFrame(fruits, columns=['Fruits']) # Columns is used to set the column name
```

```python
#Create a DataFrame from Nested Lists
people = [['Rick', 60, 'O+'], ['Monty',10,'O+'], ['summer', 45, 'A+'], ['Beth', 18, 'B+']]
df = pd.DataFrame(people, columns=['Name', 'Age', 'Blood Group'])
df
```

```python
#Create DataFrame from Dictionary
people = {'Name':['Rick', 'Monty', 'Summer', 'Rick'], 'Age':[60, 10, 45, 18], 'Blood Group':['O+', 'O+', 'A+', 'B+']}
people_df = pd.DataFrame(people)
people_df
```

```python
people_df.set_index('Name' , inplace=True)
```

```python
dictss ={"a":[10  ,11], "b":[20, 22], "c":[30, 33], "d":[40, 44], "e":[50, 55], "a":[12  ,98]}
myser=pd.DataFrame(dictss)
print(myser) # Same key is treated as index in Series but treated as column in DataFrame
```
